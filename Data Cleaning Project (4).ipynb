```python
import pandas as pd
```

```python
#importing dataset into Python
df1 = pd.read_csv(r"C:\Users\Marc\Documents\Movie Dataset2.csv")

df1
```

```python
# Dropping duplicates if any, from the dataset

df1.drop_duplicates()
```

```python
df1
```

```python
#Data cleaning process
df1['YEAR'] = df1['YEAR'].str.replace('(', '')
df1['YEAR'] = df1['YEAR'].str.replace(')', '')

df1
```

```python
df1['YEAR'] = df1['YEAR'].str.strip('V-|I')

df1['YEAR'] = df1['YEAR'].str.strip('– ')

df1
```

```python
#To remove all extra whitespaces from dataset
df1 = df1.applymap(lambda x: x.strip() if isinstance(x, str) else x)

df1
```

```python
df1['STARS'].str.strip()
```

```python
df1['STARS'] = df1['STARS'].str.replace('\n', ' ')

df1['STARS'] = df1['STARS'].str.replace('     ', '')

df1['STARS'] = df1['STARS'].str.replace('|', ', ')

df1
```

```python
df1['Gross'] = df1['Gross'].str.strip('$M')

df1
```

```python
df1 = df1.fillna('')

df1
```

```python
df1 = df1.rename(columns=lambda x: x.upper())

df1
```
